```python
from autogluon.tabular import TabularDataset, TabularPredictor
predictor = TabularPredictor.load('C:\\code\\EL\\output')
student_models = predictor.distill(time_limit=60*30)

print(student_models)
```

```python
label = 'MESFOC_nmile'
test_data = TabularDataset('C:\\code\\EL\\data\\test1.csv')
test_data_nolabel = test_data.drop(columns=[label])
preds_student = predictor.predict(test_data_nolabel, model=student_models[0])


from sklearn.metrics import r2_score
from autogluon.core.metrics import make_scorer

# 创建 r2 自定义评分器
ag_r2_scorer = make_scorer(
    name='r2',  # 评分器的名称
    score_func=r2_score,  # 使用 sklearn 的 r2_score 函数
    optimum=1,  # r2 的最佳值为 1
    greater_is_better=True  # 指定值越高越好
)
#mse评分器
def mse_func(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

ag_mse_scorer = make_scorer(
    name='mean_squared_error',
    score_func=mse_func,
    optimum=0,  # MSE的最佳值为0
    greater_is_better=True  # 越低越好
)
predictor.leaderboard(
    test_data,
    extra_metrics=[ag_mse_scorer, ag_r2_scorer]  # 包含多个自定义评分器
)
```
